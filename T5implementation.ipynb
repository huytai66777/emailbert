```python
from faker import Faker
import random

# Initialize Faker
fake = Faker()

# List of common domains
domains = [
    "example.com", "gmail.com", "yahoo.com", "outlook.com", "hotmail.com",
    "aol.com", "icloud.com", "mail.com", "zoho.com", "protonmail.com",
    "yandex.com", "gmx.com", "me.com", "live.com", "fastmail.com",
    "hushmail.com", "rocketmail.com", "rediffmail.com", "tutanota.com",
    "mail.ru", "inbox.com", "comcast.net", "verizon.net", "msn.com",
    "att.net", "sbcglobal.net", "cox.net", "earthlink.net", "btinternet.com",
    "bellsouth.net", "charter.net", "shaw.ca", "rogers.com", "telus.com",
    "xfinity.com", "optonline.net", "myway.com", "excite.com", "netscape.net",
    "mailinator.com", "temp-mail.org", "10minutemail.com"
]

# Function to generate a correct email address using Faker
def generate_correct_email():
    user_name = fake.user_name()
    domain = random.choice(domains)
    return f"{user_name}@{domain}"

# Function to introduce only correctable errors into email addresses
def introduce_correctable_error(correct_email):
    error_type = random.choice(["syntax_error", "keyboard_smash"])

    user_name, domain = correct_email.split('@')

    if error_type == "syntax_error":
        # Create an email with incorrect syntax (e.g., missing '@' or having multiple '@')
        incorrect_formats = [
            user_name + domain,              # Missing '@'
            user_name + "@.com",             # Missing domain name
            "@" + domain,                    # Missing recipient name
            user_name + "@examplecom",       # Missing top-level domain
            user_name + "@" * random.randint(2, 5) + domain  # Random number of '@'
        ]
        incorrect_email = random.choice(incorrect_formats)
        label = "syntax_error"

    elif error_type == "keyboard_smash":
        # Create a keyboard smash email address with a realistic pattern
        base_username = user_name
        smashed_part = ''.join(random.choices("asdfghjklqwertyuiopzxcvbnm1234567890", k=random.randint(5, 10)))
        smashed_user_name = f"{base_username}{smashed_part}"
        incorrect_email = f"{smashed_user_name}@{domain}"
        label = "keyboard_smash"
    return incorrect_email, label

# Function to generate the dataset
def generate_email_dataset(num_samples=1000):
    dataset = []
    for _ in range(num_samples):
        # Generate a correct email address
        correct_email = generate_correct_email()
        # Introduce an error to create an incorrect email
        incorrect_email, error_label = introduce_correctable_error(correct_email)
        # Append the pair of incorrect and correct email to the dataset
        dataset.append((incorrect_email, correct_email, error_label))
    return dataset

# Generate a dataset of 1000 email pairs
email_dataset = generate_email_dataset(10000)

# Display a few examples
for i in range(5):
    print(f"Incorrect: {email_dataset[i][0]}, Correct: {email_dataset[i][1]}, Error Type: {email_dataset[i][2]}")
```

```python
from transformers import T5ForConditionalGeneration, T5Tokenizer
from datasets import Dataset, DatasetDict
import pandas as pd

# Load the model and tokenizer
model = T5ForConditionalGeneration.from_pretrained("t5-small")
tokenizer = T5Tokenizer.from_pretrained("t5-small")

task_prefix = "email correction: "
max_length = 128

def preprocess_function(examples):
    inputs = [task_prefix + ex for ex in examples['incorrect_email']]
    targets = examples['Correct']

    # Tokenize inputs
    input_encoding = tokenizer(
        inputs,
        padding="max_length",
        max_length=max_length,
        truncation=True,
        return_tensors=None  # Return as lists
    )

    # Tokenize targets
    target_encoding = tokenizer(
        targets,
        padding="max_length",
        max_length=max_length,
        truncation=True,
        return_tensors=None  # Return as lists
    )

    # Replace padding token ids of the labels with -100
    labels = [[-100 if token_id == tokenizer.pad_token_id else token_id for token_id in seq] for seq in target_encoding.input_ids]

    # Return processed data
    return {
        'input_ids': input_encoding.input_ids,
        'attention_mask': input_encoding.attention_mask,
        'labels': target_encoding.input_ids
    }


# Convert DataFrame to Dataset
df = pd.DataFrame(df)
dataset = Dataset.from_pandas(df)

# Apply preprocessing
tokenized_dataset = dataset.map(preprocess_function, batched=True)

# Display a few examples
print(tokenized_dataset[1])
```

```python
# Assuming tokenized_dataset is a Hugging Face Dataset object
# Shuffle the dataset first
tokenized_dataset = tokenized_dataset.shuffle(seed=42)

# Split the dataset into training (70%) and temporary (30%)
split_dataset = tokenized_dataset.train_test_split(test_size=0.3, seed=42)

# Further split the temporary dataset into validation (50%) and test (50%)
temp_dataset = split_dataset['test'].train_test_split(test_size=0.5, seed=42)

# Create final splits
train_dataset = split_dataset['train']
val_dataset = temp_dataset['train']
test_dataset = temp_dataset['test']

# Create a DatasetDict
dataset_dict = DatasetDict({
    'train': train_dataset,
    'validation': val_dataset,
    'test': test_dataset
})

print(f"Training set: {len(train_dataset)} samples")
print(f"Validation set: {len(val_dataset)} samples")
print(f"Test set: {len(test_dataset)} samples")
```

```python
from transformers import Trainer, TrainingArguments, Seq2SeqTrainingArguments
import torch

batch_size = 8
model_dir = "/content/drive/MyDrive/projectdatacleaning/model1"

# Define training arguments
training_args = Seq2SeqTrainingArguments(
    model_dir,
    evaluation_strategy="steps",
    eval_steps=100,
    logging_strategy="steps",
    logging_steps=100,
    save_strategy="steps",
    save_steps=200,
    learning_rate=4e-5,
    per_device_train_batch_size=batch_size,
    per_device_eval_batch_size=batch_size,
    weight_decay=0.01,
    save_total_limit=3,
    num_train_epochs=6,
    predict_with_generate=True,
    fp16=True,
    load_best_model_at_end=True,
    report_to="tensorboard"
)
# Initialize Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset= train_dataset,
    eval_dataset= test_dataset
)

# Train the model
trainer.train()
```

```python
import torch
import numpy as np
from evaluate import load  # Use the new library as recommended

# Load metrics using the new 'evaluate' library
bleu = load("bleu")
rouge = load("rouge")

# Ensure the model is on the GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

def evaluate_model(dataset):
    exact_match_count = 0
    total_count = len(dataset)
    bleu_scores = []
    rouge_scores_1 = []
    rouge_scores_2 = []
    rouge_scores_L = []

    for item in dataset:
        incorrect = item['incorrect_email']
        correct = item['Correct']

        # Generate correction
        inputs = tokenizer(incorrect, return_tensors="pt", padding=True, truncation=True).to(device)  # Move input to device
        input_ids = inputs["input_ids"].to(device)  # Ensure input_ids are on the GPU
        outputs = model.generate(input_ids, max_length=128)
        generated = tokenizer.decode(outputs[0], skip_special_tokens=True)

        # Check exact match
        if generated.strip() == correct.strip():
            exact_match_count += 1

        # Calculate BLEU score
        bleu_score = bleu.compute(predictions=[generated.strip()], references=[[correct.strip()]])['bleu']
        bleu_scores.append(bleu_score)

        # Calculate ROUGE score
        rouge_score = rouge.compute(predictions=[generated.strip()], references=[correct.strip()])
        rouge_scores_1.append(rouge_score['rouge1'])
        rouge_scores_2.append(rouge_score['rouge2'])
        rouge_scores_L.append(rouge_score['rougeL'])

    exact_match_accuracy = exact_match_count / total_count
    avg_bleu_score = np.mean(bleu_scores)
    avg_rouge_score_1 = np.mean(rouge_scores_1)
    avg_rouge_score_2 = np.mean(rouge_scores_2)
    avg_rouge_score_L = np.mean(rouge_scores_L)

    print(f"Exact Match Accuracy: {exact_match_accuracy:.4f}")
    print(f"Average BLEU Score: {avg_bleu_score:.4f}")
    print(f"Average ROUGE-1 Score: {avg_rouge_score_1:.4f}")
    print(f"Average ROUGE-2 Score: {avg_rouge_score_2:.4f}")
    print(f"Average ROUGE-L Score: {avg_rouge_score_L:.4f}")

# Evaluate the model using the test set
evaluate_model(test_dataset)  # Evaluate a sample or subset of the test set
```
